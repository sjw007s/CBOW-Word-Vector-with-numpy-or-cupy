```python
import csv
import re

class NLPDataset(Dataset):
    def __init__(self, filenames, timesteps=10):
        super(NLPDataset, self).__init__('NLP', 'binary')
        self.training_corpus=np.array([],dtype = 'float32')
        self.corpus=np.array([],dtype = 'float32')
        self.counting_corpus = {}
        self.counting_list=[]
        self.word_to_id = {}
        self.id_to_word = {}
        self.training_one_hot_word = 0
        self.one_hot_word = 0
        self.unique_words = 0
        self.np_arange_=0
        self.negative=[]
    def text_find_number_words(self,name, training):
        full_text=list()
        name='text_data/'+name+'.txt'
        with open (name, 'rt' ,encoding='UTF8') as r:
            while True:
                line = r.readline()
                #print(1)
                if not line :
                    break
                words = line.lower().replace('.',' .').replace('\n','').split(' ')
                #print(words)
                full_text.extend(words)
                #print(full_text)
                for word in full_text:
                    if word not in self.word_to_id:
                        new_id = len(self.word_to_id)
                        self.word_to_id[word] = new_id
                        self.id_to_word[new_id] = word
            if training==1:
                self.training_one_hot_word = np.eye(len(self.word_to_id),dtype = 'float32')
                self.training_corpus = [self.word_to_id[w] for w in full_text]
                self.unique_words = len(self.word_to_id)
                for i in self.training_corpus:
                    if i not in self.counting_corpus:
                        self.counting_corpus[i] = 1
                    else:
                        self.counting_corpus[i]+=1
                        
                self.np_arange_=np.arange(len(self.word_to_id),dtype='float32')
                
                for i in range(len(self.counting_corpus)):
                    self.counting_list.append(self.counting_corpus[i]/len(full_text))
                #print(self.counting_list)
                #self.counting_list=self.counting_list/len(self.word_to_id)
                
            else:
                self.corpus = [self.word_to_id[w] for w in full_text]
            temp_len=len(full_text)
            full_text=list()
            #return temp_len, len(self.word_to_id)
        
    def create_contexts_target(self, corpus, window_size=1):
        self.target = corpus[window_size:-window_size]   #윈도우가 너무 크면 학습데이터로 사용 안된다.
        self.contexts = []
        
        for idx in range(window_size, len(corpus)-window_size):
            cs=[]
            for t in range(-window_size, window_size+1):
                if t == 0:
                    continue
                cs.append(corpus[idx + t])
            self.contexts.append(cs)
        #print(self.contexts)
        self.contexts=np.array(self.contexts,dtype='int32')
        self.target=np.array(self.target,dtype='float32')
        #print(self.counting_corpus)
        print(self.contexts.shape)
        print(self.target.shape)
    
    @property
    def train_count(self):
        return len(self.contexts)
    
    def __str__(self):
        return '{}({}, {} train_data)'.format(self.name, self.mode, self.train_count)
    
    def get_train_data(self, batch_size, nth):
        from_idx = nth * batch_size
        to_idx = (nth + 1) * batch_size

        tr_X = self.contexts[self.indices[from_idx:to_idx],:]
        tr_Y = self.target[self.indices[from_idx:to_idx]]
        
        return tr_X, tr_Y

    def shuffle_train_data(self, size):
        np.random.seed(int(time.time()))
        self.indices = np.arange(size)
        np.random.shuffle(self.indices)
        


```
